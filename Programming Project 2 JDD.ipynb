### Encoder

```python
Caesars_cipher={" ":" ","A":"D","B":"E","C":"F","D":"G","E":"H","F":"I","G":"J","H":"K","I":"L","J":"M","K":"N","L":"O","M":"P","N":"Q","O":"R","P":"X","Q":"T","R":"U","S":"V","T":"W","U":"X","V":"Y","W":"Z","X":"A","Y":"B","Z":"C",".":".","!":"!","?":"?"}
def encode(letters, Caesars_cipher):
    result=''
    for i in letters:
        if i in Caesars_cipher:
            result = result+Caesars_cipher[i]
    return result
message=input('Enter Your Message: ')
print(encode(message,Caesars_cipher))
```

### Decoder

```python
Reversed_Cipher={" ":" ","D":"A","E":"B","F":"C","G":"D","H":"E","I":"F","J":"G","K":"H","L":"I","M":"J","N":"K","O":"L","P":"M","Q":"N","R":"O","X":"P","T":"Q","U":"R","V":"S","W":"T","X":"U","Y":"V","Z":"W","A":"X","B":"Y","C":"Z",".":".","!":"!","?":"?"}
def decode(letters,Reversed_Cipher):
    result=''
    for i in letters:
        if i in Reversed_Cipher:
            result = result+Reversed_Cipher[i]
    return result
encrypted=input('Enter Your Encrypted Message: ')
print(decode(encrypted,Reversed_Cipher))
```
